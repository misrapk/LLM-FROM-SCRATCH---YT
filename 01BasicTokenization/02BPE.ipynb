```python
def get_token_pairs(text2pair):
    token_pairs = dict()

    for i in range(len(text2pair) - 1):
        pair = text2pair[i] + text2pair[i+1]
        
        
        if pair in token_pairs : 
            token_pairs[pair] += 1
        else:
            token_pairs[pair] = 1
            
    return token_pairs
```

```python
def updateVocab(token_pairs, vocab):
    mostFreqPair_idx = np.argmax(list(token_pairs.values()))
    mostFreqPair_char = list((token_pairs.keys()))[mostFreqPair_idx]
    
    vocab[mostFreqPair_char] = max(vocab.values())+1
    return vocab, mostFreqPair_char
```

```python
def generate_new_token_seq(prevText, mostFreqPair_char):
    new_text = []

    i=0
    while i<(len(prevText)-1):
        
        if (prevText[i] + prevText[i+1]) == mostFreqPair_char:
            #TODO: write some logic
            new_text.append(mostFreqPair_char)
            # print(f'added {mostFreqPair_char}')
            i+=2
        else:
            new_text.append(prevText[i])
            i+=1 
            
    if i==(len(prevText)-1):   #to append the final token
        new_text.append(prevText[i])
        
    return new_text
            
        
```

```python
text = "She had met Mr. Ranford in the grounds, and the two came back together, I remember, smiling and talking as they walked along the terrace under my window."

chars = list(set(text))
for l in chars:
  print(f'"{l}" appears {chars.count(l)} times.')
```

```python
vocab = {word:i for i, word in enumerate(chars)}
vocab
```

```python
original_text = list(text)
print(text)
print(original_text)
```

```python
vocab_size = 32

updatedText = original_text.copy()
```

```python
while len(vocab) < vocab_size:
    pairs = get_token_pairs(updatedText)
    vocab, mostFreq = updateVocab(pairs, vocab)
    updatedText = generate_new_token_seq(updatedText, mostFreq)
    print(f'Vocab has {len(vocab)} tokens after adding "{mostFreq}".')
```

```python
print(original_text)
print(updatedText)
```

```python
print(f'Original Text {original_text}')
print(f'New Text {updatedText}')
print(f'Original Token Length {len(original_text)}')
print(f'New Token Length {len(updatedText)}')
```
