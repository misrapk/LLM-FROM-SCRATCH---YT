```python
import torch
```

```python
inputs = torch.tensor(
[[0.43, 0.15, 0.89], # Your (x^1)
[0.55, 0.87, 0.66], # journey (x^2)
[0.57, 0.85, 0.64], # starts (x^3)
[0.22, 0.58, 0.33], # with (x^4)
[0.77, 0.25, 0.10], # one (x^5)
[0.05, 0.80, 0.55]] # step (x^6)
)
```

```python
#1. to calculate omeng (embedded query token)

query = inputs[1]
query
```

```python
attn_scores2 = torch.empty(inputs.shape[0])
attn_scores2
```

```python
for i, x_i in enumerate(inputs):
  attn_scores2[i] = torch.dot(x_i, query)


attn_scores2
```

```python
#2. normalize each of the attention scores

attn_weight2 = attn_scores2/attn_scores2.sum()

print("Attention Weights: ", attn_weight2)
print("Attention weight sum : ", attn_weight2.sum())
```

```python

#using exponential funtion
attn_weight2_exp = torch.exp(attn_scores2)/torch.exp(attn_scores2).sum()

attn_weight2_exp
```

```python
attn_weight2_exp.sum()
```

```python
#3. to calcualte context vector

attn_weight2[1] *inputs[1]
```

```python
context_vec2 = torch.zeros(query.shape)
for i, x_i in enumerate(inputs):
  context_vec2 += attn_weight2[i] *x_i

print(context_vec2)
```
